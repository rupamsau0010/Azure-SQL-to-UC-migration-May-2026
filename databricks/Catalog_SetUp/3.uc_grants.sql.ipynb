```sql
-- ============================================================
-- Catalog level — all groups need USE CATALOG to enter
-- ============================================================
GRANT USE CATALOG ON CATALOG migration_dev
  TO `grp-data-engineers`;

GRANT USE CATALOG ON CATALOG migration_dev
  TO `grp-data-analysts`;

-- ============================================================
-- Bronze schema — engineers read/write, analysts no access
-- ============================================================
GRANT USE SCHEMA, CREATE TABLE, SELECT, MODIFY
  ON SCHEMA migration_dev.bronze
  TO `grp-data-engineers`;

-- ============================================================
-- Silver schema — engineers read/write, analysts no access
-- (analysts get aggregated Gold, not raw Silver)
-- ============================================================
GRANT USE SCHEMA, CREATE TABLE, SELECT, MODIFY
  ON SCHEMA migration_dev.silver
  TO `grp-data-engineers`;

-- ============================================================
-- Gold schema — engineers full access, analysts read-only
-- ============================================================
GRANT USE SCHEMA, CREATE TABLE, SELECT, MODIFY
  ON SCHEMA migration_dev.gold
  TO `grp-data-engineers`;

GRANT USE SCHEMA, SELECT
  ON SCHEMA migration_dev.gold
  TO `grp-data-analysts`;

-- ============================================================
-- Quarantine — engineers only (contains PII-adjacent bad data)
-- ============================================================
GRANT USE SCHEMA, CREATE TABLE, SELECT, MODIFY
  ON SCHEMA migration_dev.quarantine
  TO `grp-data-engineers`;

-- ============================================================
-- External Locations — engineers only
-- ============================================================
GRANT READ FILES, WRITE FILES
  ON EXTERNAL LOCATION ext_loc_bronze
  TO `grp-data-engineers`;

GRANT READ FILES, WRITE FILES
  ON EXTERNAL LOCATION ext_loc_processed
  TO `grp-data-engineers`;

GRANT READ FILES, WRITE FILES
  ON EXTERNAL LOCATION ext_loc_quarantine
  TO `grp-data-engineers`;

-- ============================================================
-- Verify grants
-- ============================================================
SHOW GRANTS ON SCHEMA migration_dev.gold;
SHOW GRANTS ON EXTERNAL LOCATION ext_loc_bronze;
```
