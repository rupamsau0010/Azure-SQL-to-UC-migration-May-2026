```python
# Create catalog
spark.sql("CREATE CATALOG IF NOT EXISTS migration_dev")
spark.sql("USE CATALOG migration_dev")

# Create schemas
for schema in ["bronze", "silver", "gold", "quarantine"]:
    spark.sql(f"CREATE SCHEMA IF NOT EXISTS migration_dev.{schema}")
    print(f"Created schema: {schema}")

# Verify
display(spark.sql("SHOW SCHEMAS IN migration_dev"))
```
